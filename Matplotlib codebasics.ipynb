```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 46, 48, 46, 40, 42, 41]
```

```python
# to set the axis label
plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('Weather')

plt.plot(days, max_t, label = 'MAX')
plt.plot(days, min_t, label = 'MIN')
plt.plot(days, avg_t, label = 'AVG')

# after giving label for the blue green orange lines also it wont show

# so put legend funciton at last

plt.legend()

# change the location of legend to
# bottom left or bottom right or upper right or upper left

plt.legend(loc = 'upper left')

# by default, the location is set 'best'
plt.legend(loc = 'best', shadow = True, fontsize = 'small')

# shadow effect on legend

plt.grid()

# grid for lines 
```

# BAR CHART

```python
company = ['Google', 'Amazon', 'Microsoft', 'FaceBook']
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
ypos = np.arange(len(company))
```

```python
ypos
```

```python
plt.xticks(ypos, company) 
plt.xlabel('Reveue(bin)')
plt.title('US Tech stocks')
# len is (0, 1, 2, 3)
# xticks replaces len by com names like google, amazon ...
plt.bar(ypos-0.2, revenue, width = 0.4, label = 'Revenue')
plt.bar(ypos+0.2, profit, width = 0.4, label = 'Profit')
plt.legend()
```

```python
xpos = np.arange(len(company))
```

```python
xpos
```

```python
# Horizontal bar chart
# use barh

plt.yticks(ypos, company) 
plt.xlabel('Reveue(bin)')
plt.title('US Tech stocks')
# len is (0, 1, 2, 3)
# xticks replaces len by com names like google, amazon ...
plt.barh(xpos-0.2, revenue, label = 'Revenue')
plt.barh(xpos+0.2, profit, label = 'Profit')
plt.legend()
```

# Histograms

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82]
blood_sugar_women = [67, 45, 98, 120, 133, 150, 84, 69, 89, 120, 112]
```

```python
plt.xlabel('Sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men,blood_sugar_women], bins = [80, 100, 125, 150], rwidth = 0.95, color = ['green', 'orange'], label = ['men', 'women'], orientation = 'horizontal')
plt.legend()
```

# Pie Chart

```python
exp_values = [1400, 600, 300, 410, 250]
exp_labels = ['Home rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilities']
```

```python
plt.axis('equal')
plt.pie(exp_values, labels = exp_labels, radius = 1.5, autopct = '%0.1f%%', shadow = True, explode=[0, 0.5, 0, 0, 0], startangle = 180)
plt.show()
```
