```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_temp = [50,51,52,48,47,49,46]
min_temp = [43,42,40,44,33,35,37]
avg_temp = [45,48,48,46,40,42,41]
```

# Axes labels and chart title

```python
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Weather")

plt.plot(days, max_temp, label = "Max Temp")
plt.plot(days, min_temp, label = "Min Temp")
plt.plot(days, avg_temp, label = "Avg Temp")
```

# Legend

```python
#  Show legend

plt.plot(days, max_temp, label = "Max Temp")
plt.plot(days, min_temp, label = "Min Temp")
plt.plot(days, avg_temp, label = "Avg Temp")

plt.legend(loc = "best")
```

```python
# legend at different location with shadow enabled 

plt.plot(days, max_temp, label = "Max Temp")
plt.plot(days, min_temp, label = "Min Temp")
plt.plot(days, avg_temp, label = "Avg Temp")

plt.legend(loc = 'upper right', shadow = True, fontsize = 'large')

plt.grid()
```
